### Importando Bibliotecas

```python
## importando bibliotecas necess√°rias para o funcionamento da atividade
# Pandas: Ferramentas de an√°lise e manipula√ß√£o de dados
# Sklearn: Biblioteca de Machine Learn e que tem a base de dados dos digitos
# Numpy: Biblioteca para computa√ß√£o num√©rica
# Matplotlib: Biblioteca para plotagem de gr√°ficos


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

from sklearn import datasets, svm, metrics
digits = datasets.load_digits()

# Importando dados e respostas para vari√°veis X e y
X,y = load_digits(return_X_y=True)

```

### Criando dataframe com Pandas para visualiza√ß√£o

```python
# criando dataframe
df = pd.DataFrame(data = digits.data, columns = digits.feature_names)

# adicionando coluna de respota no dataframe
df['numero'] = digits.target

# retorna 10 primeiras linhas
df.head(10)
```

### Segmentando base

```python
# treinamento e teste (teste=30% e treinamento = 70%)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=13)
```

### Encontrando o melhor K

```python
# criando loop de 2 a 50 para encontrar o melhor k poss√≠vel

i = 2
k_resutados = []

while i <= 50:

  n_neighbors = i

  clf = KNeighborsClassifier(n_neighbors=n_neighbors)
  clf.fit(X_train, y_train)

  y_pred = clf.predict(X_test)

  accuracy = accuracy_score(y_test, y_pred)
  k_resutados.append([i,accuracy]) 
  i+=1

print(k_resutados)
```

```python
# gr√°fico de k

data = np.array(k_resutados)
x, y = data.T
plt.plot(x,y)
plt.title('Melhor Acur√°cia por K')
plt.xlabel('K')
plt.ylabel('Acur√°cia')
plt.show()
```

```python
k, max_value= max(k_resutados, key=lambda item: item[1])

print('A maior acur√°cia √©:', max_value, "com o K igual √†:",k)
```

```python
n_neighbors = k

clf = KNeighborsClassifier(n_neighbors=n_neighbors)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
k_resutados.append([i,accuracy]) 
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy*100,"%")
```

### Analisando os resultados obtidos

```python
import seaborn as sns
matrix=confusion_matrix(y_test,y_pred)
sns.heatmap(matrix, square = False, annot = True, cbar=False)
plt.xlabel('Classifica√ß√£o Predita')
plt.ylabel('Classifica√ß√£o Correta')
plt.suptitle("Matriz de Confus√£o - KNN")
```
